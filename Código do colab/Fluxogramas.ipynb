```python
#Recebe o preco de fabrica
custo_fabrica = float(input('Digite o custo de fabrica do carro: R$ '))
```

```python
#Calcula o valor final
impostos = custo_fabrica*0.45
valor_com_impostos = custo_fabrica+impostos
distribuicao = valor_com_impostos*0.28
custo_final = valor_com_impostos+distribuicao
```

```python
#Recebe o resultado
print(f"\nCusto ao consumidor: r$ {custo_final:.2f}")
```

horaextra.drawio.png

```python
#Calculo de pagamento com horas extras
print("Inicio")
```

```python
#Entrada de dados
horas = 16
minutos = 30
valor_hora = 30.0
```

```python
#Calculo do total de horas
total_horas = horas+(minutos/60)
print (f"total = horas + (min/60) resultado = {total_horas}")
```

```python
#Calculo do pagamento base
pagamento_base = total_horas*valor_hora
print(f"pagamento = total*valor resultado = {pagamento_base} + 70% de h/extra")
```

```python
#Calculo do pagamento com 70% extra
pagamento_total = pagamento_base*1.7
print(f"pagamento = ${pagamento_total:.2f}")
```
