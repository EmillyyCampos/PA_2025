#Numero primo

```python
numero = int(input("Digite um número: "))
eh_primo = numero > 1 and all(numero % i != 0 for i in range(2, numero))

if eh_primo: print(f"{numero} é um número primo!")
else:
    print(f"{numero} não é um número primo.")
```

#Calculadora simples

```python
# Entrada
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Processamento
operacao = input("Digite a operação (+, -, *, /): ")

if operacao == "+":
    resultado = num1 + num2
    print(f"Resultado: {resultado}")
else:
    if operacao == "-":
        resultado = num1 - num2
        print(f"Resultado: {resultado}")
    else:
        if operacao == "*":
            resultado = num1 * num2
            print(f"Resultado: {resultado}")
        else:
            if operacao == "/":
                if num2 != 0:
                    resultado = num1 / num2
                    print(f"Resultado: {resultado}")
                else:
                    print("Erro: divisão por zero.")
            else:
                print("Operação inválida.")
```

#Jogo de adivinhação

```python
import random

# Gera um número aleatório entre 1 e 10
numero_aleatorio = random.randint(1, 10)

#Entrada do usuário
palpite = int(input("Tente adivinhar o número (entre 1 a 10): "))

# Processamento e saída
if palpite == numero_aleatorio:
  print("Parabéns! Você acertou!")
else:
  print(f"Que pena! O número correto era {numero_aleatorio}. Tente novamente!")
```

#Número positivo, negativo ou zero

```python
numero = float(input("Digite um número: "))

if numero > 0:
    print("O número é positivo.")
else:
    if numero < 0:
        print("O número é negativo.")
    else:
        print("O número é zero.")
```

#Calculo IMC

```python
# Entrada
peso = float(input("Digite seu peso em kg: "))
altura = float(input("Digite sua altura em metros: "))

# Processamento
imc = peso / (altura ** 2)

# Saída
print(f"Seu IMC é: {imc:.2f}")
```

#Calculo da media

```python
#entrada
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))

#processamento
media = (nota1 + nota2 + nota3) / 3

if media >= 6:
  print("Você foi aprovado!")
else:
   print("Você está de recuperação.")
```
